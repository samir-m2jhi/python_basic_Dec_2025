Guessing a number game.

```python
import random

print("Guess a number between 1-100")
c = random.randrange(1, 101)
attempts = 0
max_attempts = 5

while attempts < max_attempts:
    i = int(input(f"Attempt {attempts + 1}: Enter your guess: "))
    attempts += 1

    if i < c:
        print("Your guess is low")
    elif i > c:
        print("Your guess is high")
    else:
        print(f" Your guess is correct! You guessed it in {attempts} attempts.")
        break
else:
    # This executes if the while loop is exhausted (user didn't guess correctly)
    print(f" Sorry! You've used all {max_attempts} attempts. The correct number was {c}.")
```

Rock, Paper, and Scissor Game

```python
import random
l=['rock','paper','scissor']
print("You will be 5 chances for each player!")
user_score=0
attempt=0
c_score=0
while attempt<=5:
  print(f"Round {attempt}")
  attempt+=1
  ur=int(input('''
  Game start.....
  1) Yes
  2) No | Exit

  '''))
  if ur==1:
    for a in range(1,6):
      i=int(input('''
    1) Rock
    2) paper
    3) Scissor
      '''))
      if i==1:
        u='rock'
      elif i==2:
        u='paper'
      elif i==3:
        u='scissor'
      else:
        print('Invalid Input choose number 1-3 only!')
      cchoise=random.choice(l)
      print(f'Your choise: {u}')
      print(f'Computer Choice:{cchoise}')
      if u==cchoise:
        print('Draw')
      else:
        if (u=='rock' and cchoise=='paper') or (u=='paper' and cchoise == 'scissor') \
        or (u=='scissor' and cchoise=='rock'):
          print('Computer Win')
          c_score+=1
        else:
          print('You Win')
          user_score+=1

  else:
      print('Game Over')
      break
print("Game Finished!")
print(f"Your Score:{user_score}")
print(f"Computer Score:{c_score}")
if user_score>c_score:
  print("You Won the Game")
elif user_score<c_score:
  print("Computer Won the Game")
else:
  print("Game is Draw")
```
